```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as pl
```

```python
df = pd.read_csv('funda-amsterdam.csv')
df.head()
```

```python
#creating a subset to exclude the implausible data 
df_subset2 = df[df['price']>30000]
df_subset2 = df_subset2[['area', 'rooms', 'price']]
```

```python
#scatterplotmaxtric of the three quantitative variables
sns.pairplot(df_subset2)
```

```python
#scatterplot one combination to find out if there is a lineair combination
sns.scatterplot(data=df_subset2, x = 'price', y = 'rooms')
```

Yes it is a linair correlation, when the price is higher --> the amount of rooms is highter 

```python
#calculating the pearson' r
from scipy.stats.stats import pearsonr #We will need the correlation function from the scientific package scipy

pearsonr(df_subset2['price'],df_subset2['rooms'])
```

The correlation is moderate, it is between 0.3 and 0.6
