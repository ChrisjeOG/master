```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('surveydata.csv')
df1 = pd.read_csv('stepsdata.csv', sep=';')
df.head()
```

```python
df1.head()
```

```python
df3 = pd.merge(df1, df, on = 'id')
```

```python
df3.head()
```

```python
df3['waist'].value_counts()
```

```python
df3['waist'] = df3['waist'].apply(pd.to_numeric, errors='coerce')
def above_45(x):
    if(x < 45.0): 
        return float('NaN')
    else: 
        return x
df3['waist'] = df3['waist'].apply(above_45)

df3.plot(y='waist')
```
